## Preliminary

<font size='4'>
    
* There are six files stored under `Your own GitHub repository/data/Quiz4`.
* Import those six files using relevant functions in `Pandas`.

```python
df1 = pd.read_csv("data/Quiz4/data_1.txt")
df2 = pd.read_csv("data/Quiz4/data_2.txt")
df3 = pd.read_csv("data/Quiz4/data_3.txt")
df4 = pd.read_csv("data/Quiz4/data_4.txt")
df5 = pd.read_csv("data/Quiz4/data_5.txt")
df6 = pd.read_csv("data/Quiz4/data_6.txt")
dfs = [df1, df2, df3, df4, df5, df6]
# dfs
```

## Import Data

<font size='4'>

* Three lists named `study_id_ls`, `measure_date_ls`, and `duration_ls` have been initialized.
* Write a for loop to extract and append the values of `study_id`, `measure_date`, and `duration` to `study_id_ls`, `measure_date_ls`, and `duration_ls`, respectively.

```python
study_id_ls = []
measure_date_ls = []
duration_ls = []
for df in dfs:
    study_id_ls.append(df.study_id.iloc[0])
    measure_date_ls.append(df.measure_date.iloc[0])
    duration_ls.append(df.duration.iloc[0])
# print(study_id_ls)
# print(measure_date_ls)
# print(duration_ls)
```

<font size='4'>

* Create a new Pandas Dataframe combining three lists and name it `input_df`.

```python
input_df = pd.DataFrame({
    "study_id": study_id_ls,
    "measure_date": measure_date_ls,
    "duration": duration_ls
})
# input_df
```

<font size='4'>

* Visualize `input_df`

```python
input_df
```

## Data Manipulation

<font size='4'>

* Reorder the dataframe by `measure_date` (earliest to latest) and then by `duration` (shortest to longest).

```python
input_df = input_df.sort_values(by=["measure_date", "duration"], ascending=[True, True])
# input_df
```

<font size='4'>

* Create a new column called `relative_time` by calculating the time difference in days between each of the measure date and the earliest one.

```python
input_df["measure_date"] = pd.to_datetime(input_df["measure_date"])
earliest = input_df["measure_date"].min()
input_df["relative_time"] = (input_df["measure_date"] - earliest).dt.days
# input_df
```

<font size='4'>

* Reset the index column and remove the old index.

```python
input_df = input_df.reset_index(drop=True)
```
